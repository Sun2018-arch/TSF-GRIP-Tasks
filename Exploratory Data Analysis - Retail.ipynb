# As a business manager, try to find out the weak areas where you can work to make more profit?

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

Technology and Office Supplies have high profit whereas Furniture have less profit

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

Major states to concentrate sales are California, New York, Texas, Washington, Pennsylivania, Florida where Prfits shall be increased to realise gains

Improvements are to be done in same day ship mode. More work to be done in Southern region for better business. Work to be done in furniture and technology category as well. Copier Recruitment must increase.Profit and sales must be made such that they remain directly proportional with each other.
